## 1. Probability Distribution

```python
#import necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import pyplot
from numpy import mean,std
from scipy.stats import norm
#Select Data File
path='c:/pydatafiles/usa.csv'
Data=pd.read_csv(path)
#Select a data sample
Sample1=Data.head()
print(Sample1)
```

```python
#Drop column "Day" as it may not be useful in predicting...
Data.drop(["Day"],axis=1,inplace=True)
Data
```

# Frequency Distribution

```python
#Histogram for frequency distribution of Deaths
print('Frequency Distribution')
pyplot.title("Frequency Distribution for Deaths")
pyplot.xlabel("Deaths")
pyplot.ylabel("Frequency")
plt.hist(Data['Deaths'], bins=20)
```

```python
#Estimating the parameters for the density function
D_mean=mean(Data['Deaths'])
D_std=std(Data['Deaths'])
print('Deaths Mean: ',D_mean)
print('\nDeaths Standard deviation:',D_std)
```

# Normal Probability Distribution

```python
#Choose the range of values in the data set on which to plot PDF
min_Deaths=min(Data['Deaths'])
max_Deaths=max(Data['Deaths'])
values=list(range(int(min_Deaths), int(max_Deaths)))
print(values)
```

```python
#Create pdf objects by passing mean and std parameters
Gaussian_pdf_object=norm(D_mean,D_std)
```

```python
#Convert frequencies into probabilities
probabilities=[Gaussian_pdf_object.pdf(value) for value in values]
print("\n Probabilities\n",probabilities)
```

```python
#Plot Normal Probability Distribution Curve
print('Normal probabilities distribution Curve')
pyplot.title("Normal probabilities distribution Deaths")
pyplot.xlabel("Deaths")
pyplot.ylabel("Normal probabilities")
pyplot.plot(values, probabilities)
```

```python
#This curve is bell-shaped, meaning that the data is normally distributed
#The line plot shows the Probability Density Function for a Normal Distribution
#The highest probability (0.00016) is around the mean (3251.09)
```

```python
plt.hist(Data['Deaths'], bins=20, density=True)
pyplot.plot(values, probabilities)
```

### Cumulative Distribution

```python
#plot cumulative probabilities of normal(gaussian) distribution of Deaths
#i. use Gaussian pdf_object to create a Gaussian cumulative density function object.
Gaussian_cpdf_object = [Gaussian_pdf_object.cdf(value) for value in values]
#ii.use the Gaussian cumulative density function object to plot cumulative probability# or the Cumulative Density Function for the Normal Distribution
pyplot.plot(values, Gaussian_cpdf_object)
print('Cumulative probabilities distribution Curve')
pyplot.title("Cumulative Density function weights")
pyplot.xlabel("Deaths")
pyplot.ylabel("cumulative probabilities")
pyplot.show()
#
#The plot shows that the probability of observing the mean Deaths (3251.09) and below
# is 0.5(50%) That is, the there is 50% probability for observing mean Deaths and below
```

# 2. Bayessian Classifier

```python
#import necessary libraries
import numpy as np
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import accuracy_score
import pandas as pd
from sklearn import preprocessing
from sklearn.preprocessing import Normalizer
from sklearn.decomposition import PCA
```

```python
#Select dataset
path='C:/pydatafiles/mexico.csv'
Data=pd.read_csv(path)
print(Data)
```

```python
#Pre-process the Data
#i.Transform Target Population to Numeric array
Encoder=preprocessing.LabelEncoder()
Encoded_Data=Data.apply(preprocessing.LabelEncoder().fit_transform)
print("Transformed Data:\n",Encoded_Data)
Numeric_Array=Encoded_Data.values
print("Numeric Array\n",Numeric_Array)
#ii. Select Samples
#a. select training sample
split=10
Training_Sample=Numeric_Array[:split,:]
print("Training Sample:\n",Training_Sample)
#b. Select Test Sample
Test_Sample=Numeric_Array[split:,:]
print("Test Sample\n",Test_Sample)
#c.Select Independent Attributes Sample
#1. Select independent attributes of training sample
X_Train_Sample=Training_Sample[:,:-1]
print("Independent Attributes of Training sample\n", X_Train_Sample)
#2. Select Dependent attribute of Training sample
Y_Train_Sample=Training_Sample[:,-1]
print("Dependent Attribute of Training Sample\n\n",Y_Train_Sample)
#3. Select Independent attributes of Test Sample
X_Test_Sample= Test_Sample[:,:-1]
print("Independent Attributes of Test Sample\n\n",X_Test_Sample)
Y_Test_Sample= Test_Sample[:,-1]
print("dependent Attribute of Test Sample\n\n",Y_Test_Sample)
```

```python
# Bayessian Classification
#i. Train the Naive Bayes algorithm
#Create a Gaussian Classifier object
Classifier_object = GaussianNB()
#ii. Train the classifier object by passing X_training and Y_Train as parameters
model=Classifier_object.fit(X_Train_Sample,Y_Train_Sample)
#Using The Model to predict label of unknown instance
#iii. Predict Output by passing X_Test sample to predict method as parameter
predicted_y=model.predict(X_Test_Sample)
print("Predicted Value:", predicted_y)
```

```python
accuracy=accuracy_score(Y_Test_Sample,predicted_y)
print("Accuracy rate:",accuracy*100)
```

### The model is 53.67% accurate in predicting the likelihood of a confirmed case having pre-existing medical conditions.
